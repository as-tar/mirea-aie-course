```python
import numpy as np
```

```python
a1 = np.array([1, 2, 3]) # одномерный массив
a1
```

```python
type(a1) # тип массива
```

```python
a1.dtype # тип данных элементов массива
```

```python
a1.shape # форма массива
```

```python
a2 = np.array([[1, 2, 3], [4, 5, 6]]) # двумерный массив
a2
```

```python
a2.shape # 2 строки, 3 столбца
```

```python
np.arange(0, -1, -0.25) # последовательность чисел из диапазона от 0 до -1 (не включительно) с шагом -0.25
```

```python
np.zeros((3, 5)) # двумерный массив из нулей
```

```python
np.ones((5, 3)) # двумерный массив из единиц
```

```python
np.eye(5) # единичная матрица
```

```python
a1[0] # получение отдельного элемента массива
```

```python
a1[0:3:2] # извлечение подмассива с помощью среза
```

```python
b1 = np.array([10, 20, 30])
b2 = np.array([[11, 22, 33], [44, 55, 66]])
```

```python
a1 + b1 # поэлементное сложение одномерных массивов
```

```python
a2 + b2 # поэлементное сложение двумерных массивов
```

```python
a1 * b1 # поэлементное умножение одномерных массивов
```

```python
a2 * b2 # поэлементное умножение двумерных массивов
```

```python
c = np.array([[10, 100, 1000], [20, 200, 2000], [30, 300, 3000]])
c
```

```python
c.sum() # суммирование всех элементов массива
```

```python
c.min() # минимальный элемент массива
```

```python
c.max() # максимальный элемент массива
```

```python
c.mean() # среднее значение всех элементов массива
```
